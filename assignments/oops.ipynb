## Objective
This notebook demonstrates all the main OOPS concepts using minimum examples from 3 classes each for Encapsulation, Inheritance, Polymorphism, and Extraction.  


## 1) Polymorphism

```python
class Animal:                               # class 1
    def sound(self):
        return "Animal makes sound"

class Dog(Animal):                          # class 2
    def sound(self):
        return "Dog barks"
        
class Cat(Animal):                          # class 3
    def sound(self):
        return "Cat meows"

a1 = Animal()
d1 = Dog()
c1 = Cat()

print("Polymorphism\n")
for obj in [a1, d1, c1]:
    print(obj.sound())
```

## 2) Abstraction

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):                         # Abstract class
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):                         # class 1
    def start(self):        
        return "Car hereeeee"

class Bike(Vehicle):                        # class 2
    def start(self):
        return "Bike hereeeeeee"

class Bus(Vehicle):                         # class 3
    def start(self):
        return "Bus hereeee"

c1 = Car()
b1 = Bike()
bs1 = Bus()

print("Abstraction\n")
print(c1.start())
print(b1.start())
print(bs1.start())
```

## 3) Inheritance

```python
class Animal:                             #Parent class
    def eat(self):
        return "Animal here...."

class Dog(Animal):                        # class 1
    def bark(self):
        return "Dog"

class Cat(Animal):                        # class 2
    def meow(self):
        return "Cat"

class Lion(Animal):                       # class 3
    def roar(self):
        return "Lion"

d = Dog()
c = Cat()
l = Lion()

print("Inheritance\n")
print(d.eat(), "and", d.bark())
print(c.eat(), "and", c.meow())
print(l.eat(), "and", l.roar())
```

## 4) Encapsulation

```python
class BankAccount:                        # class 1
    def __init__(self, balance):
        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

    def get_balance(self):
        return self.__balance


class Student:                            # class 2
    def __init__(self, name, marks):
        self.name = name
        self.__marks = marks

    def set_marks(self, marks):
        self.__marks = marks

    def get_marks(self):
        return self.__marks


class Employee:                           # class 3
    def __init__(self, salary):
        self.__salary = salary

    def increase_salary(self, amt):
        self.__salary += amt

    def get_salary(self):
        return self.__salary

acc = BankAccount(1000)
stu = Student("Aman", 85)
emp = Employee(30000)

acc.deposit(500)
stu.set_marks(90)
emp.increase_salary(5000)

print("Encapsulation\n")
print(acc.get_balance())
print(stu.get_marks())
print(emp.get_salary())
```
