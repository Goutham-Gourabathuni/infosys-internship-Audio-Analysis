## Input Example

```python
from IPython.display import display
import ipywidgets as widgets

name_widget = widgets.Text(
    value='',
    placeholder='Enter your name',
    description='Name:',
    disabled=False
)
submit = widgets.Button(description='Submit')
out = widgets.Output()

def on_submit(b):
    with out:
        out.clear_output()
        print('Hello', name_widget.value)

submit.on_click(on_submit)
display(name_widget, submit, out)
```

```python
# The parent class holds the common logic
class speechtotextmodel:
    def __init__(self):
        # The child classes should override this setup_model method
        self.model = self.setup_model() 

    # This method is common to all
    def transcribe(self, audio_file):
        # We assume self.model is correctly initialized by the child class
        return self.model.transcribe(audio_file)

# The child class only focuses on what makes it unique (loading the specific model)
class whisperstt(speechtotextmodel):
    def setup_model(self):
        # Load the specific Whisper model implementation
        return load_whisper_model() 

class googlestt(speechtotextmodel):
    def setup_model(self):
        # Load the specific Google STT implementation
        return load_google_stt_client()
```
